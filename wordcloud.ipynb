```python
import pymongo
from pymongo import MongoClient

import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

import gensim
from gensim import corpora
from pprint import pprint
```

```python
# creating a client to retrieve the data
client = MongoClient()
```

```python
# getting the database
ba_database = client['ba_database']

# getting the collection
ba_col = ba_database['ba_review_col']
```

```python
# collecting the clean tokens of all the reviews from the database

review_tokens = []
for tokens in ba_col.find({}, {'_id':0, 'tokens':1}):
        review_tokens.append(tokens['tokens'])
```

```python
# joining the tokens into a long sentence
for tokens in review_tokens:
        text = " ".join(word for word in tokens)
```

```python
# creating a wordcloud
word_cloud = WordCloud(collocations = False, background_color = '#010101').generate(text)
```

```python
# displaying the wordcloud
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
